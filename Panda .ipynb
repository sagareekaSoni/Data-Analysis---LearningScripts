## Indexing 

Index based selection
* First row then column (This is the opposite of what we do in native Python, which is column-first, row-second)
* iloc : on its numeric positions 


```python
wine_reviews.iloc[0]
```

```python
wine_reviews.iloc[[1,2],1]
```

Label-Based Selection
 * .loc: the data index value, not its position

```python
wine_reviews.loc[0,"country"]
```

```python
wine_reviews.loc[0,:]
```

* iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded.So 0:10 will select entries 0,...,9. 
* loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.

### Maps

there are two mapping method:
* map() : should expect a single value from the Series
* apply() : the equivalent method if we want to transform a whole DataFrame

```python
meanreview=wine_reviews.points.mean()
wine_reviews.points.map(lambda a:a - meanreview)
```

```python
def remean_points(row):
    row.points = row.points - meanreview
    return row
```

```python
wine_reviews.apply(remean_points, axis='columns')
```

## Topic 4 : Grouping  and Sorting

```python
# Group by 
#Groupby() created a group of COUNTRY COUNT.
wine_reviews.groupby('country').country.count().head()
```

```python
# First row of each slice group 
wine_reviews.groupby('country').apply(lambda df: df.winery.iloc[0]).head() #head rows
```

```python
wine_reviews.groupby(['country', 'winery']).apply(lambda df: df.loc[df.points.idxmax()])
```
